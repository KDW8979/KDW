```python
import pandas as pd
from wordcloud import WordCloud
import matplotlib.pyplot as plt
from collections import Counter
import re

#  
file_path = r'C:\Users\kdp\Desktop\KDW\EX_Deep_Learning\data\MBTI 500.csv'
data = pd.read_csv(file_path)

# MBTI  'P'   
p_type_data = data[data['type'].str.contains('P')]

#      
all_posts = ' '.join(p_type_data['posts'].values)

#       
cleaned_text = re.sub(r'[^a-zA-Z\s]', '', all_posts).lower()

#   
words = cleaned_text.split()

#    
word_counts = Counter(words)

#  30  
top_40_words = [word for word, count in word_counts.most_common(40)]

#  10      
filtered_word_counts = {word: count for word, count in word_counts.items() if word not in top_40_words}

#       200  
most_common_words = dict(Counter(filtered_word_counts).most_common(200))

#  
wordcloud = WordCloud(width=800, height=400, background_color='white', colormap='inferno').generate_from_frequencies(most_common_words)

#  
plt.figure(figsize=(10, 5))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.show()

```
