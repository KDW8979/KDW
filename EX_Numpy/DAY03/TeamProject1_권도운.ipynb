```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import koreanize_matplotlib
```

```python
df=pd.read_excel(r'C:\Users\kdp\Desktop\KDW\EX_Numpy\DAY03\상품성과_2024-01-01_2024-06-30.xlsx')
df.head()
```

```python
df=df[['상품명','환불금액','환불비율(결제금액)','환불수량','환불건수','환불비율(결제상품수량)']]
df
```

```python
df1=df.sort_values(['환불건수','환불금액'], ascending=False)
df1
```

```python
df1=df1.head(10)
df1
```

```python
x=df1['상품명']
y=df1['환불건수']
plt.figure(figsize=(10,10))
plt.barh(x,y)
plt.xlabel('환불건수')
plt.ylabel('상품명')
plt.title('가장 많은 환불건수 상품 TOP 10',fontsize=20)
plt.gca().invert_yaxis()
plt.show()
```

```python
df2=df.sort_values('환불수량', ascending=False)
df2=df2.head(10)
df2
```

```python
x=df2['상품명']
y=df2['환불수량']
plt.figure(figsize=(10,10))
plt.barh(x,y)
plt.xlabel('환불수량')
plt.ylabel('상품명')
plt.title('가장 많은 환불수량 상품 TOP 10',fontsize=20)
plt.gca().invert_yaxis()
plt.show()
```
