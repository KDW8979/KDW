```python
# [1] 모듈 로딩
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# [2] 데이터 준비
FILE_DATA=r'C:\Users\kdp\Desktop\KDW\EX_PANDAS06\DATA\남북한발전전력량.xlsx'
```

```python
# [3] EXCEL => DataFrame형태로 저장
dataDF=pd.read_excel(FILE_DATA)
```

```python
# [4] 데이터 확인
dataDF.info()
print(dataDF.head())
```

```python
# [5] 데이터 전처리 및 가공
# - 멀티인덱싱 : 컬럼 ==> 행인덱스 설정
# - (1) NaN컬럼에 대한 처리
# - (2) 2개의 컬럼을 행인덱스로 설정
```

```python
# - (1) NaN컬럼에 대한 처리
dataDF[dataDF.columns[0]]       # dataDF['전력량(억kWh)']
```

```python
# dataDF[dataDF.columns[0]].fillna(method='ffill')
dataDF[dataDF.columns[0]].ffill(inplace=True)
```

```python
dataDF.head()
```

```python
# 행인덱스 사용할 컬럼명 추출
dataDF.columns[:2].to_list()
```

```python
# - (2) 2개의 컬럼을 행인덱스로 설정

dataDF1=dataDF.set_index(dataDF.columns[:2].to_list(), inplace=True)
print(f'[확인]\n{dataDF.head(2)}')
```

```python
idx_names=dataDF.index.names
print(f'idx_name : {idx_names}')

dataDF.index.set_names({idx_names[0] : '구분', idx_names[1] : '전력'}, inplace=True)
print(f'[확인]\n{dataDF.head()}')
```

```python
# - '-' 데이터값 <NA>으로 치환
dataDF.replace('-',pd.NA, inplace=True)
print(f'[확인\n{dataDF.head()}')
```

```python
dataDF.isna().T.sum()
```

```python
dataDF.fillna(0,inplace=True)
dataDF.isna().T.sum()
```

```python
# - object ==> int16 형변환
dataDF=dataDF.astype('int16')
dataDF.dtypes
```

```python
# [6] 남한과 북한 발전량 합계 비교
plt.plot(dataDF.loc['남한','합계'], 'bo-', label='남한-합계')
plt.plot(dataDF.loc['북한','합계'], 'ro-', label='북한-합계')
plt.title('[남한&북한 발전량]')
plt.xlabel('Year')
plt.ylabel('Value')
plt.legend()
plt.show()
```

```python
def drawLineGraph(title,columns,labels,xlabel,ylabel):
    for col in columns:
        plt.plot(dataDF.loc[col])
    # for idx, col in enumerate(columns):
    #     plt.plot(dataDF.loc[col], 'bo-', label=f'{labels[idx]}')            
    plt.title(f'---[{title}]---')
    plt.xlabel('Year')
    plt.ylabel('Value')
    plt.legend(labels)
    plt.show()
```

```python
# - figure 객체 생성
# - 생성된 figure 객체 안에 subplot 객체 생성
#   * 그래프를 표시 방식 지정 : 가로/세로 몇개 그래프 넣을 지 지정
# - subplot에 그래프 넣기
```

```python
# -> 하나의 Figure에서 수력/화력/원자력/합계 그래프 넣기
fig=plt.figure() 

# -add_subplot(행수,열수,번호)
ax1=fig.add_subplot(2,2,1)
ax2=fig.add_subplot(2,2,2)
ax3=fig.add_subplot(2,2,3)
ax4=fig.add_subplot(2,2,4)

# - 각 subplot에 그래프 넣기
ax1.plot(dataDF.loc['남한','수력'], 'bo-', label='남한-수력')
ax2.plot(dataDF.loc['남한','화력'], 'bo-', label='남한-화력')
ax3.plot(dataDF.loc['남한','원자력'], 'bo-', label='남한-원자력')
ax4.plot(dataDF.loc['남한','합계'], 'bo-', label='남한-합계')

# - 각 subplot별로 설정
ax1.set_title('남한 수력발전량')
ax1.set_xlabel('연도')
ax2.set_title('남한 화력발전량')
ax2.set_xlabel('연도')
ax3.set_title('남한 원자력발전량')
ax3.set_xlabel('연도')
ax4.set_title('남한 합계발전량')
ax4.set_xlabel('연도')


# - 화면에 띄우기
plt.tight_layout()
plt.legend()
plt.show()
```

```python
# -> 하나의 Figure에서 수력/화력/원자력/합계 그래프 넣기
fig=plt.figure(figsize=(14,8)) 

# - subplot 객제 생성 : sbplots() => axes 객체를 배열에 담아서 반환
#                                => 행단위로 axes 담아서 반환
axes=fig.subplots(1,4)


# - subplot에 그래프 넣기
datas=dataDF.index.to_list()[:4]
fmfs=['bo-','go-','ro-','ko--']
for idx,ax in enumerate(axes):
    ax.plot(dataDF.loc[datas[idx]],fmfs[idx])

# - 전체 창 즉, figure 설정
fig.suptitle("남한의 에너지 발전량 변화 동향", fontsize=12) 
    
# - 화면에 띄우기
plt.tight_layout()
plt.show()
```
