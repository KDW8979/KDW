- [5] 분석 : 대구에서 전지역 전입 인구수 분석

```python
# [5-1] 대구에서 타지역 전입 인구수 분석

cityDF
```

```python
cityDF['합계']=0
cityDF['합계']=cityDF.loc[:,'1970':'2017'].sum(axis=1)
cityDF
```

```python
cityDF1=cityDF['합계']
cityDF1
```

```python
# 그래프 그리기
cities1=cityDF1.iloc[:7]
cities2=cityDF1.iloc[7:]
```

```python
# 한글폰트
from matplotlib import font_manager as fm, rc


Font_File=r'C:\Windows\Fonts\malgun.ttf'


font_name=fm.FontProperties(fname=Font_File).get_name()


rc('font',family=font_name)
```

```python
fig=plt.figure(figsize=(15,15))

ax1=fig.add_subplot(1,2,1)
ax2=fig.add_subplot(1,2,2)
ax1.bar(cities1.index,cities1.values, color=['r','b'])
ax2.bar(cities2.index,cities2.values, color=['r','b'])
plt.tight_layout()
plt.show()
```
