### 연도별 경제 성장률, 실업률 등 경제 지표와 범죄 발생률 간의 관계를 분석
### 경제 상황 변화 시 범죄 발생률의 변동 패턴을 분석하여 경제적 요인이 범죄에 미치는 영향을 평가
#### 2020~2022

```python
import pandas as pd
```

```python
# 연도별 경제 성장률
# 2021년 : 4.3%, 2022년 : 2.6%, 2023년 : 1.4%     갈수록 성장률이 낮아지고 있다.

```

```python
# 연도별 실업률
file_name='행정구역_시도_실업률.xlsx'
```

```python
unempDF=pd.read_excel(file_name)
unempDF
```

```python
unempDF=unempDF.drop(columns='연령계층별',axis=1)
unempDF
```

```python
unempDF=unempDF.set_index('시도별')
unempDF
```

```python
unempDF.index=['서울','부산','대구','인천','광주','대전','울산','세종','경기','강원','충북','충남','전북','전남','경북','경남','제주']
unempDF
```

```python
unempDF.columns=['2020실업률','2021실업률','2022실업률']
unempDF
```

```python
unempDF['2021실업률증강률']=unempDF['2021실업률']-unempDF['2020실업률']
unempDF
```

```python
unempDF['2022실업률증강률']=unempDF['2022실업률']-unempDF['2021실업률']
unempDF
```

```python
unempDF1=unempDF[['2021실업률증강률','2022실업률증강률']]
unempDF1
```

```python
# 지역별 범죄율 정리
file_name3='범죄발생지역.xlsx'
```

```python
chDF=pd.read_excel(file_name3)
chDF
```

```python
chDF=chDF.drop(columns='범죄별(1)',axis=1)
chDF
```

```python
chDF=chDF.rename(columns={'발생지역별(1)':'시도별'})
chDF
```

```python
chDF=chDF.set_index('시도별')
chDF
```

```python
chDF.columns=['2020범죄율','2021범죄율','2022범죄율']
chDF
```

```python
chDF['2021범죄율증강량']=chDF['2021범죄율']-chDF['2020범죄율']
chDF
```

```python
chDF['2022범죄율증강량']=chDF['2022범죄율']-chDF['2021범죄율']
chDF
```

```python
chDF1=chDF[['2021범죄율증강량','2022범죄율증강량']]
chDF1
```

```python
result=pd.concat([chDF,unempDF],axis=1)
print(result)
```

```python
# 범죄율증강량, 실업률증강률만 뽑아보자
result2=pd.concat([chDF1,unempDF1],axis=1)
result2
```

```python
# 지역별 GDP 정리
file_name4='시도별gdp..xlsx'
```

```python
gdpDF=pd.read_excel(file_name4)
gdpDF
```

```python
gdpDF=gdpDF.drop('경제활동별',axis=1)
gdpDF
```

```python
gdpDF=gdpDF.drop(0)
gdpDF
```

```python
gdpDF=gdpDF.set_index('시도별')
gdpDF
```

```python
gdpDF.index=['서울','부산','대구','인천','광주','대전','울산','세종','경기','강원','충북','충남','전북','전남','경북','경남','제주']
gdpDF
```

```python
gdpDF.columns=['2020GDP','2021GDP','2022GDP']
gdpDF
```

```python
gdpDF['2021GDP증강량']=gdpDF['2021GDP']-gdpDF['2020GDP']
gdpDF
```

```python
gdpDF['2022GDP증강량']=gdpDF['2022GDP']-gdpDF['2021GDP']
gdpDF
```

```python
gdpDF1=gdpDF[['2021GDP증강량','2022GDP증강량']]
gdpDF1
```

```python
result3=pd.concat([chDF,gdpDF],axis=1)
result3
```

```python
# 범죄율증강량, 실업률증강률만 뽑아보자
result4=pd.concat([chDF1,gdpDF1],axis=1)
result4
```
