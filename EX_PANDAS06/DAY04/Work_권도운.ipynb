```python
import pandas as pd
```

```python
file_name='titanic.csv'
```

```python
ttDF=pd.read_csv(file_name)
ttDF
```

```python
ttDF.info()
```

```python
print(ttDF.head(),ttDF.tail(), sep='\n\n')
```

```python
# 빈칸 확인
ttDF.isna().sum()
```

```python
# 그냥 확인
ttDF.describe()
```

```python
# survived열과 alive열은 중복되므로 survived열 삭제
ttDF.drop(['survived'],axis=1, inplace=True)
ttDF
```

```python
# sex열과 who열이 중복되므로 who열 삭제
ttDF.drop(['who'],axis=1, inplace=True)
ttDF
```

```python
# embarked열과 embark_town도 중복되므로 embark열 삭제
ttDF.drop(['embarked'],axis=1, inplace=True)
ttDF
```

```python
# pclass열과 class열도 중복되니 pclass 삭제
ttDF.drop(['pclass'],axis=1,inplace=True)
ttDF
```

```python
# adult_male열은 age와 sex열을 보면 되므로 가차없이 삭제
ttDF.drop(['adult_male'],axis=1,inplace=True)
```

```python
# embark_town의 빈 칸은 가장 많이 나오는 값으로 넣어버리기
ttDF['embark_town'].mode()
```

```python
ttDF['embark_town']=ttDF['embark_town'].fillna('Southampton')
```

```python
ttDF.isna().sum()
```

```python
# 나이 빈칸은 대충 평균으로 맞춰넣기
print(ttDF['age'].median())
ttDF['age'].fillna(ttDF['age'].median(),inplace=True)
ttDF
```

```python
# deck열도 빈칸이 너무 많아서 그냥 버리기로 함
ttDF.drop(['deck'],axis=1,inplace=True)
ttDF
```

```python
# age열과 fare열은 정수로 바꾸기로 함
ttDF['age']=ttDF['age'].astype('int64')
ttDF['fare']=ttDF['fare'].astype('int64')
ttDF
```

```python
# sibsp열과 parch열은 묶어서 가족으로 퉁치자
familyDF=ttDF['sibsp']+ttDF['parch']
familyDF
```

```python
ttDF['family']=familyDF
ttDF
```

```python
ttDF.drop(['sibsp'],axis=1,inplace=True)
ttDF.drop(['parch'],axis=1,inplace=True)
ttDF
```

```python
# 질문 1. 살아남은 사람들 중 가족이 4명 이상 같이 탔던 사람의 수는?
answer=(ttDF['alive']=='yes') & (ttDF['family']>=4)
ttDF[answer].count()[0]
```

```python
# 질문 2. 1등석에 탄 성인 여성들 중 살아남은 사람의 수는?
answer2=(ttDF['class']=='First')&(ttDF['age']>=20)&(ttDF['sex']=='female')&(ttDF['alive']=='yes')
ttDF[answer2].count()[0]
```

```python
# 질문 3. 3등석에 탄 성인 남성들 중 살아남은 사람의 수는?
answer3=(ttDF['class']=='Third')&(ttDF['age']>=20)&(ttDF['sex']=='male')&(ttDF['alive']=='yes')
ttDF[answer3].count()[0]
```
