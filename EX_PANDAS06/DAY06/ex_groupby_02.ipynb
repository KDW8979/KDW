#### [그룹화 처리]
- 주제 : 타이타닉에서 선실에 등급에 따른 생존 비율을 분석

```python
import pandas as pd
```

```python
file1='titanic.csv'
```

```python
titanicDF=pd.read_csv(file1)
titanicDF
```

```python
pclass_aliveDF=titanicDF[['pclass','alive']]
pclass_aliveDF
```

```python
pclass_aliveDF.isna().sum()
```

```python
groupObj=pclass_aliveDF.groupby('pclass')
```

```python
print(groupObj.groups.keys())
print(groupObj.groups.values())
```

```python
groupObj.count()
```

```python
result=groupObj.value_counts()
result.index
```
