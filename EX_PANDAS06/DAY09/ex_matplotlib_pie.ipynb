#### [ 파이 그래프 ]
- 컬럼(변수)의 분포 비율을 시각화

```python
# [1] 모듈 로딩
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
# [2] 데이터 준비
F_DATA=r'C:\Users\kdp\Desktop\KDW\EX_PANDAS06\DATA\auto_mpg.csv'
```

```python
# [3] CSV => DataFrame으로 저장
dataDF=pd.read_csv(F_DATA)
dataDF
```

```python
# - 컬럼별 타입 화긴
dataDF.dtypes
```

```python
# [4] 특정 컬럼의 데이터 비율
# - 자동차 생산국가에 대한 비율
dataDF['origin'].astype('category')
```

```python
# - 자동차 생산국가에 대한 비율 
round(dataDF['origin'].value_counts()/dataDF['origin'].count(),2)
```

```python
# - 자동차 생산국가에 대한 비율 시각화 - bar 그래프
originSR=dataDF['origin'].value_counts()
x=originSR.index.to_list()
h=originSR.values.tolist()
x,h
```

```python
originBar=plt.bar(x,h, color=['purple','skyblue','grey'], edgecolor=['crimson'])
# 막대에 라벨을 설정
plt.bar_label(originBar, h, label_type='center')

# x축 눈금 설정
plt.xticks([1,2,3],['US','EU','JP'], rotation=-45, fontsize=15)

plt.show()
```

```python
# - 자동차 생산국가에 대한 비율 시각화 - pie 그래프
plt.pie(h, labels=['US','EU','JP'], autopct='%.2f%%')

plt.show()
```

```python
# - 자동차 생산국가에 대한 비율 시각화 - pie 그래프
plt.pie(h, labels=['US','EU','JP'], autopct='%.2f%%', counterclock=False, startangle=90, explode=[1, 0.5, 0])

plt.show()
```

```python
# - 자동차 생산국가에 대한 비율 시각화 - pie 그래프
plt.pie(h, labels=['US','EU','JP'], autopct='%.2f%%', counterclock=False, startangle=90, 
        wedgeprops={'linewidth':1, 'width':0.5, 'edgecolor':'gold'})

plt.show()
```
